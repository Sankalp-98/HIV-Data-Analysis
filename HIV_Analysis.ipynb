```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_excel('HIV_0000000006.xlsx')
df1
```

```python
for i in range(len(df1)):
    df1['Number of deaths due to HIV/AIDS'][i]=df1['Number of deaths due to HIV/AIDS'][i].split("[")[0]
```

```python
df1
```

```python
for j in range(len(df1)):
    l=df1['Number of deaths due to HIV/AIDS'][j].split(" ")
    ans=""
    for i in range(len(l)):
        ans+=l[i]
    df1['Number of deaths due to HIV/AIDS'][j]=int(ans)
```

```python
df1
```

```python
df1.to_excel('Updated_HIV_0000000006.xlsx',index=False)
```
